RNN(Recurrent Neural Network)은 순환 신경망으로, 시퀀스 데이터를 처리하는 데 특화된 딥러닝 모델이다. 일반적인 신경망과 달리 RNN은 이전 시점의 정보를 기억하고, 이를 현재 예측에 반영할 수 있는 구조를 가지고 있다.

### RNN의 주요 특징:
1. **순환 구조**: RNN은 네트워크가 시간적으로 연결되어 있어, 이전 단계의 정보를 활용할 수 있다. 즉, 현재 상태를 계산할 때 이전의 히든 상태(hidden state)를 참고하여 시계열 데이터나 순차적 데이터 처리에 적합하다.
   
2. **시간 의존성**: RNN은 시계열 데이터, 예를 들어 자연어 처리(NLP), 음성 인식, 주식 가격 예측 등에서 시간에 따른 데이터의 흐름을 학습할 수 있다.

3. **장기 의존성 문제**: RNN은 기본적으로 이전의 정보를 전달하지만, 긴 시퀀스를 처리할 때는 '기울기 소실(vanishing gradient)' 문제로 인해 과거의 정보가 잘 전달되지 않는 한계가 있다. 이를 개선하기 위해 LSTM(Long Short-Term Memory)이나 GRU(Gated Recurrent Unit) 같은 변형 모델이 자주 사용된다.

RNN의 이러한 특성 덕분에 시계열이나 순차적인 데이터 분석에서 매우 유용하게 쓰인다.
