```python
# Hugging Face Transformers 라이브러리 설치
!pip install transformers

# Hugging Face의 GPT2 모델과 토크나이저 불러오기
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# GPT2 모델 및 토크나이저 로드
model_name = "gpt2"  # gpt2 대신 gpt2-medium, gpt2-large, gpt2-xl 등의 버전을 사용할 수 있음
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# 프롬프트 입력
prompt = "In the future, artificial intelligence will"

# 입력 텍스트를 토큰화하여 모델에 입력
inputs = tokenizer.encode(prompt, return_tensors="pt")

# 텍스트 생성
outputs = model.generate(
    inputs,
    max_length=100,   # 생성할 텍스트의 최대 길이
    num_return_sequences=1,  # 생성할 문장의 개수
    no_repeat_ngram_size=2,  # 반복 방지를 위한 n-gram 크기
    do_sample=True,  # 샘플링을 통해 창의적인 텍스트 생성
    top_k=50,        # top-k 샘플링으로 가능한 50개 중 선택
    top_p=0.95,      # top-p 샘플링으로 누적 확률 95% 내에서 선택
    temperature=0.7  # 출력의 창의성 조절 (0.0 ~ 1.0)
)

# 생성된 텍스트 디코딩 및 출력
generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(generated_text)
```

문장 점수 평가

```python
# Hugging Face Transformers 라이브러리 설치
!pip install transformers

# 필요한 라이브러리 임포트
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# GPT2 모델과 토크나이저 로드
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# 평가 함수 정의
def evaluate_sentence(prompt, user_sentence):
    # 프롬프트와 사용자 문장을 합치기
    full_prompt = f"{prompt} {user_sentence}"

    # 입력 텍스트를 토큰화하여 모델에 입력
    inputs = tokenizer.encode(full_prompt, return_tensors="pt")

    # 텍스트 생성
    outputs = model.generate(
        inputs,
        max_length=len(inputs[0]) + 30,  # 생성할 최대 길이
        num_return_sequences=1,
        no_repeat_ngram_size=2,
        do_sample=True,
        top_k=50,
        top_p=0.95,
        temperature=0.7
    )

    # 생성된 텍스트 디코딩
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # 점수 매기기 (예: 단어 수 기준으로 간단히 점수 매기기)
    score = len(generated_text.split())

    return generated_text, score
```

```python
# 사용자 입력 프롬프트
initial_prompt = "It was a dark and stormy night."
print("초기 문장:", initial_prompt)

# 사용자가 한 문장을 입력받음
user_sentence = input("다음 문장을 입력하세요: ")

# 문장 평가
generated, score = evaluate_sentence(initial_prompt, user_sentence)

# 결과 출력
print("\n생성된 문장:", generated)
print("점수:", score)
```
